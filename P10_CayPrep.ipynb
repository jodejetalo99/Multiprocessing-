```python
import learn2clean.loading.reader as rd 
import pandas as pd

# ejecutamos la función de creación de perfiles para un conjunto de datos como entrada
rd.profile_summary(datos, plot=False)
```

```python
# codificación de la variable objetivo, que en este caso usé la de "Resultado" pues
# Aquí se indica si se padece o no de coronavirus

d_enc = rd.Reader(sep=',',verbose=True, encoding=True) 

datos_encoded = d_enc.train_test_split(['/home/jodejetalo99/200415COVID19MEXICO.csv'], 'RESULTADO')
datos_encoded['train'].head()
datos_encoded['test'].head()
```

```python
# Selección de atributos

import  learn2clean.feature_selection.feature_selector  as fs

fs.Feature_selector(dataset = datos_encoded.copy(), strategy= 'IMP', exclude = 'SibSp',verbose=True , threshold =.4).transform()
```

```python
import learn2clean.normalization.normalizer as nl 

# Normalización MM con exclusión = Ninguna, todas las variables numéricas se normalizarán
n1= nl.Normalizer(datos_encoded.copy(),strategy='MM',exclude='RESULTADO')

n1.transform()['train'].head()
```

```python
# algoritmos para limpiar datos faltantes
import learn2clean.imputation.imputer as imp
# Reemplazamos los valores faltantes por los más frecuentes en los conjuntos de datos de capacitación y prueba

imp1 = imp.Imputer(datos_encoded.copy(),strategy='MF', verbose=True).transform()
imp2 = imp.Imputer(datos_encoded.copy(),strategy='MEDIAN', verbose=True).transform()
imp1['train'].head()
```

```python
# Detectar valores atípicos y los eliminamos
import learn2clean.outlier_detection.outlier_detector as out
#to remove rows having 40% and more ZSB-based outling values among the numerical variables; 
# since LOF requires non missing values, rows with NaN are also removed
out3=out.Outlier_detector(datos_encoded.copy(), strategy='LOF', threshold = .4, verbose=False)
out3.transform()['train'].head()
```

```python
# Eliminacion de registros duplicados

import learn2clean.duplicate_detection.duplicate_detector as dup

#Remove approximate duplicates with thresholding Jaccard similarity 
# using 'AD'strategy of the Duplicate_detector class
dup2 = dup.Duplicate_detector(datos_encoded, strategy='AD', threshold = .6, verbose=True).transform()

dup2['train']
```

```python
# Ahora permita que Lear2Clean con qlearner establezca el pipeline y modelo
# creamos la tubería de preprocesamiento para la clasificación
import learn2clean.loading.reader as rd 
import learn2clean.normalization.normalizer as nl 
import learn2clean.feature_selection.feature_selector as fs
import learn2clean.duplicate_detection.duplicate_detector as dd
import learn2clean.outlier_detection.outlier_detector as od
import learn2clean.imputation.imputer as imp
import learn2clean.classification.classifier as cl
d_enc = rd.Reader(sep=',',verbose=True, encoding=True) 
datos_encoded = d_enc.train_test_split(['/home/jodejetalo99/200415COVID19MEXICO.csv'], 'RESULTADO')
# reemplazamos valores numéricos faltantes por mediana
d1 = imp.Imputer(dataset=datos_encoded, strategy = 'MEDIAN',verbose=False).transform()
# escala decimal para variables numéricas
d2 = nl.Normalizer(dataset=d1, strategy='DS', exclude = 'Sentiment', verbose=False).transform()
# eliminamos 20 LOF outliers
d3 = od.Outlier_detector(dataset=d2, strategy='LOF', threshold= 0.2,verbose=False).transform()
# clasificamos con LDA
cl.Classifier(dataset=d3,strategy = 'LDA', target = 'RESULTADO', verbose =True).transform()
```
